```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from somoclu import Somoclu

df = pd.read_csv('../data/rings_data.csv', index_col=0)
df.head()
```

```python
sns.pairplot(data=df, hue='label')
plt.show()
```

```python
rings_X = df[['x','y','z']].values
rings_label = df.label.values

som = Somoclu(n_rows=15, n_columns=25)
som.train(rings_X)
som.view_umatrix(labels=rings_label, bestmatches=True, filename="som_rings.png")
```
