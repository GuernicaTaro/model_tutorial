```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from somoclu import Somoclu
import matplotlib.pyplot as plt
```

```python
# loading data, reducing its dimension by PCA
digits = load_digits()
pca = PCA(n_components=0.95)
pca_data = pca.fit_transform(digits.data)
print(pca_data.shape)
```

```python
# データ数が多いので間引く
np.random.seed(0)
sample_index = np.random.choice(pca_data.shape[0], 800, replace=False)
```

```python
# somで可視化
sample_X = pca_data[sample_index]
sample_y = digits.target[sample_index]

som = Somoclu(n_rows=200, n_columns=200, initialization="pca")
som.train(sample_X)
som.view_umatrix(labels=sample_y, bestmatches=True, filename="som_digits.png")
```
