```python
import pandas as pd

# graphics
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d    # <- projection='3d'
%matplotlib inline

from somoclu import Somoclu

from sklearn.datasets import make_swiss_roll
num = 300
noise = 0.01
X, t = make_swiss_roll(num, noise)
label = (0.3*t).astype(int)      # 0.3*tintcast

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(X[:, 0], X[:, 1], X[:, 2],
                s=5,                    # 
                c=label,                # 
                cmap=plt.cm.Spectral)   # 
plt.colorbar(sc)                        # 
```

```python
som = Somoclu(n_rows=40, n_columns=80)
som.train(X)
som.view_umatrix(labels=label, bestmatches=True, filename="som_swissroll.png")
```
