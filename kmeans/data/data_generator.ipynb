```python
import pandas as pd
import numpy as np

np.random.seed(0)
```

```python
n1 = 20
s1_x = pd.Series(np.random.normal(0,1,n1), name='x')
s1_y = pd.Series(np.random.normal(-1,1,n1), name='y')
df1 = pd.concat([s1_x,s1_y], axis=1)
df1['label'] = 0
```

```python
n2 = 30
s2_x = pd.Series(np.random.normal(3,1,n2), name='x')
s2_y = pd.Series(np.random.normal(1,1,n2), name='y')
df2 = pd.concat([s2_x,s2_y], axis=1)
df2['label'] = 1
```

```python
n3 = 20
s3_x = pd.Series(np.random.normal(6,1,n3), name='x')
s3_y = pd.Series(np.random.normal(-1,1,n3), name='y')
df3 = pd.concat([s3_x,s3_y], axis=1)
df3['label'] = 2
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.pairplot(data=df, hue='label')
plt.show()
```

```python
df.to_csv('data.csv')
```
